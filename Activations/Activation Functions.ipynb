```python
import numpy as np
import matplotlib.pyplot as plt
```

**ReLU - Rectified Linear unit<br>**
<br>
ReLU is the most common activation function used in neural networks it is defined as:<br>
$ f(x) = max(0,x)$
<br>
or we could define it as:
$$f(x) = \begin{cases} 
1, & x > 0 \\
0, & x \leq 0
\end{cases}
$$

```python
def ReLU(n):
    "Returns 0 is input is negative and returns the same number if greater or equal to zero"
    return np.maximum(0,n)
```

```python
# sample data to test the relu activation
x = np.arange(-3,5)
x
```

```python
A = ReLU(x)
A
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.plot(x)
plt.title("Before applying ReLU")


plt.subplot(1,2,2)
plt.plot(ReLU(x))
plt.title("After Applying ReLU")

plt.tight_layout()
plt.show()
```
