```python
import numpy as np
import matplotlib.pyplot as plt
```

### ReLU - Rectified Linear unit<br>
<br>
ReLU is the most common activation function used in neural networks it is defined as:<br>

$$
f(x) = max(0,x)
$$

or we could define it as:

$$
\text{ReLU}(x) =
\begin{cases} 
    x, & \text{if } x > 0 \\
    0, & \text{if } x \leq 0
\end{cases}
$$

```python
def ReLU(n):
    "Returns 0 is input is negative and returns the same number if greater or equal to zero"
    return np.maximum(0,n)
```

```python
# sample data to test the relu activation
x = np.arange(-15,15)
x
```

```python
A = ReLU(x)
A
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.plot(x)
plt.title("Before applying ReLU")


plt.subplot(1,2,2)
plt.plot(ReLU(x))
plt.title("After Applying ReLU")

plt.tight_layout()
plt.show()
```

### Sigmoid
Sigmoid function is defined as follows
$$
f(x) = \frac{1}{1 + e^{-x}}
$$
It squashes the input in the range[0,1]

```python
def sigmoid(n):
    return 1/ (1+ np.exp(-n))

S = sigmoid(x)
S
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x)
plt.title("Before Sigmoid")

plt.subplot(1,2,2)
plt.plot(S)
plt.title("After Sigmoid")
```

### Tan Hyperbolic
Tanh is defined as follows:
$$
f(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}
$$
It squashes the input in the range [-1,1]

```python
def tanh(n):
    numerator = np.exp(n) - np.exp(-n)
    denominator = np.exp(n) + np.exp(-n)
    return numerator/denominator
```

```python
T = tanh(x)
T
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x)
plt.title("Before Tanh")

plt.subplot(1,2,2)
plt.plot(T)
plt.title("After Tanh")
```

## Leaky ReLU
Leaky ReLU is similar to ReLU, but for negative values, it returns a small portion (usually \( \frac{1}{100} \)) of the input, avoiding dead neurons.
$$
f(x) = \max(0.01 \cdot x, x)
$$
Alternatively, we can define it as:

$$
\text{LeakyReLU}(x) =
\begin{cases} 
    x, & \text{if } x > 0 \\
    0.01x, & \text{if } x \leq 0
\end{cases}
$$




```python
def leakyReLU(n):
    return np.maximum(0.01*n,n)
```

```python
x = np.arange(-30,20,3)
```

```python
L = leakyReLU(x)
L
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x)
plt.title("Before LeakyReLU")

plt.subplot(1,2,2)
plt.plot(L)
plt.title("After LeakyReLU")
```

We can barely see this in the graph on the right but the negative values have been mutiplied by 0.01 and aren't zero like ReLU
