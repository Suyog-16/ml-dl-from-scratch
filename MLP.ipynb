### Mathematical fourmulation

- **Hidden Layer**
$$
z_1 = x_1w_{11} + x_2w_{21} + b_1

$$
$$

z_2 = x_1w_{12} + x_2w_{22} + b_1

$$

Apply ReLU

$$ a_1 = max(0,z_1) $$
$$ a_2 = max(0,z_2) $$


- **Ouput Layer**

$$ z_3 = a_1w_3 + a_2w_4 + b_3 $$

Output
$$  \hat{y} = \sigma{(z_3)}  $$

Where
we sigmoid is as :
$$ \hat{y} = \frac{1}{1+e^{-z}}  $$



**Loss function**: Binary Cross Entropy

$$ L = - [y log(\hat{y})+ (1-y)log(1-\hat{y})]$$


### Activation function



```python
def relu(x):
    return np.maximum(0,x)

def relu_derivative(x):
    return ( x >0 ).astype(float)

def sigmoid(x):
    return (1 / (1 +np.exp(x) ))
```

### Dataset

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],[1],[1],[0]])
```

### Intialize Parameters

```python
import numpy as np

np.random.seed(42)

w11 = np.random.randn() * 0.1
w12 = np.random.randn() * 0.1
w21 = np.random.randn() * 0.1
w22 = np.random.randn() * 0.1

w3 = np.random.randn() * 0.1
w4 = np.random.randn() * 0.1

b1 = 0.0
b2 = 0.0
b3 = 0.0
```

### Single Forward Propagation

```python
# Input
x1 = 1
x2 = 0

# Hidden Layer
z1 = x1 * w11 + x2 * w21 + b1
a1 = relu(z1)

z2 = x1 * w12 + x2 * w22 + b2
a2 = relu(z2)

# Output Layer
z3 = a1 * w3 + a2 * w4 + b3
y_hat = sigmoid(z3)
```

```python
loss = -(y * np.log(y_hat + 1e-8) +(1 - y) * np.log(1 - y_hat + 1e-8)) # 1e-8 to avoid zero
```
