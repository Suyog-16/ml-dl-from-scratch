### Mathematical fourmulation

- **Hidden Layer**
$$
z_1 = x_1w_{11} + x_2w_{21} + b_1

$$
$$

z_2 = x_1w_{12} + x_2w_{22} + b_1

$$

Apply ReLU

$$ a_1 = max(0,z_1) $$
$$ a_2 = max(0,z_2) $$


- **Ouput Layer**

$$ z_3 = a_1w_3 + a_2w_4 + b_3 $$

Output
$$  \hat{y} = \sigma{(z_3)}  $$

Where
we sigmoid is as :
$$ \hat{y} = \frac{1}{1+e^{-z}}  $$



**Loss function**: Binary Cross Entropy

$$ L = - [y log(\hat{y})+ (1-y)log(1-\hat{y})]$$


### Activation function



```python
def relu(x):
    return np.maximum(0,x)

def relu_derivative(x):
    return ( x >0 ).astype(float)

def sigmoid(x):
    return (1 / (1 +np.exp(x) ))
```
