### Mathematical fourmulation

- **Hidden Layer**
$$
z_1 = x_1w_{11} + x_2w_{21} + b_1

$$
$$

z_2 = x_1w_{12} + x_2w_{22} + b_1

$$

Apply ReLU

$$ a_1 = max(0,z_1) $$
$$ a_2 = max(0,z_2) $$


- **Ouput Layer**

$$ z_3 = a_1w_3 + a_2w_4 + b_3 $$

Output
$$  \hat{y} = \sigma{(z_3)}  $$

Where
we sigmoid is as :
$$ \hat{y} = \frac{1}{1+e^{-z}}  $$



**Loss function**: Binary Cross Entropy

$$ L = - [y log(\hat{y})+ (1-y)log(1-\hat{y})]$$


### Activation function



```python
def relu(x):
    return np.maximum(0,x)

def relu_derivative(z):
    return np.where(z > 0, 1, 0)

def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

### Dataset

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],[1],[1],[0]])
```

### Intialize Parameters

```python
import numpy as np

np.random.seed(42)

w11 = np.random.randn() * 1.0
w12 = np.random.randn() * 1.0
w21 = np.random.randn() * 1.0
w22 = np.random.randn() * 1.0

w3 = np.random.randn() * 1.0
w4 = np.random.randn() * 1.0

b1 = np.random.randn() * 0.1
b2 = np.random.randn() * 0.1
b3 = np.random.randn() * 0.1
```

### Single Forward Propagation

```python
# Input
x1 = 1
x2 = 0

# Hidden Layer
z1 = x1 * w11 + x2 * w21 + b1
a1 = relu(z1)

z2 = x1 * w12 + x2 * w22 + b2
a2 = relu(z2)

# Output Layer
z3 = a1 * w3 + a2 * w4 + b3
y_hat = sigmoid(z3)
```

```python
loss = -(y * np.log(y_hat + 1e-8) +(1 - y) * np.log(1 - y_hat + 1e-8)) # 1e-8 to avoid zero
```

## Backpropagation

```python
# Output Layer
dz3 = y_hat - y

dw3 = a1 * dz3
dw4 = a2 * dz3
db3 = dz3

# Hidden Layer
da1 = dz3 * w3
da2 = dz3 * w4

dz1 = da1 * relu_derivative(z1)
dz2 = da2 * relu_derivative(z2)

dw11 = x1 * dz1
dw21 = x2 * dz1
db1 = dz1

dw12 = x1 * dz2
dw22 = x2 * dz2
db2 = dz2
```

## Gradient Descent

```python
learning_rate = 0.01

w11 -= learning_rate * dw11
w21 -= learning_rate * dw21
b1  -= learning_rate * db1

w12 -= learning_rate * dw12
w22 -= learning_rate * dw22
b2  -= learning_rate * db2

w3 -= learning_rate * dw3
w4 -= learning_rate * dw4
b3 -= learning_rate * db3
```

## Building Functions

```python
def loss_fn(y,y_hat):
    return -(y * np.log(y_hat + 1e-8) +(1 - y) * np.log(1 - y_hat + 1e-8))

def forward(x1,x2):
    z1 = x1 * w11 + x2 * w21 + b1
    a1 = relu(z1)

    z2 = x1 * w12 + x2 * w22 + b2
    a2 = relu(z2)

    z3 = a1 * w3 + a2 * w4 + b3
    y_hat = sigmoid(z3)

    return z1,a1,z2,a2,z3,y_hat

def backward(w3,w4,x1, x2, y, z1, a1, z2, a2, z3, y_hat):
    
    dz3 = y_hat - y

    dw3 = a1 * dz3
    dw4 = a2 * dz3
    db3 = dz3

    # Hidden Layer
    da1 = dz3 * w3
    da2 = dz3 * w4

    dz1 = da1 * relu_derivative(z1)
    dz2 = da2 * relu_derivative(z2)

    dw11 = x1 * dz1
    dw21 = x2 * dz1
    db1 = dz1

    dw12 = x1 * dz2
    dw22 = x2 * dz2
    db2 = dz2

    return (
        dw11, dw12,
        dw21, dw22,
        dw3, dw4,
        db1, db2, db3
    )

```

```python
def optimizer(
    w11, w12, w21, w22,
    w3, w4,
    b1, b2, b3,

    dw11, dw12,
    dw21, dw22,
    dw3, dw4,
    db1, db2, db3,

    learning_rate
):

    w11 -= learning_rate * dw11
    w12 -= learning_rate * dw12
    w21 -= learning_rate * dw21
    w22 -= learning_rate * dw22

    w3 -= learning_rate * dw3
    w4 -= learning_rate * dw4

    b1 -= learning_rate * db1
    b2 -= learning_rate * db2
    b3 -= learning_rate * db3

    return (
        w11, w12, w21, w22,
        w3, w4,
        b1, b2, b3
    )
```

```python
epochs = 1000
learning_rate = 0.01
for epoch in range(epochs):
    total_loss = 0

    for i in range(len(X)):
        x1,x2 = X[i]
        
        y_label = y[i][0]

        # forward pass
        z1,a1,z2,a2,z3,y_hat = forward(x1,x2)

        # compute loss

        loss = loss_fn(y_label,y_hat)
        total_loss += loss

        # backprop
        gradients = backward(
            w3, w4, x1, x2, y_label,
            z1, a1,
            z2, a2,
            z3, y_hat
        )


        (
            w11, w12, w21, w22,
            w3, w4,
            b1, b2, b3
        ) = optimizer(
            w11, w12, w21, w22,
            w3, w4,
            b1, b2, b3,

            *gradients,

            learning_rate
        )
    if epoch % 100 == 0:
        print(f"Epoch {epoch} | Loss = {total_loss / len(X):.4f}")
```
