# Naive Bayes

Naive Bayes is a **probablistic generative model** based on the Bayes Theorem with a assumption that all features are independent to each other . 

## Mathematical Fourmulation

Let 

$ X = \{ x_1,x_2,x_3,....,x_n \}$  be a feature vectors with all the features

$
C \in \{c_1,c_2,c_3,.....c_n\} $ be class labels


We want to model :

$$
P(C | X)
$$




## Bayes Theorem


We want to compute the probaility of a class $C$ given the data distribution of features $X$

$$
P(C | X ) = \frac{P(X|C). P(C)}{P(X)}
$$


Where:

- $P(C | X)$ - Posterior ( What we need to find out)
- $P(X | C)$ - Likelihood
- $P(C)$ - Priors
- $P(X)$ - Evidence

$P(X)$ is constant across classes

## Naive Conditional Independece Assumption

Without Conditional Indpendence Assumption(**True Form**):
$$
P(X| C_k ) = P(x_1,x_2,..,x_n | C_k)

$$
This is a **joint distribution** which assumes that all features depend on each other
$$
P(X∣C_k​)=P(x_1​∣x_2​,…,x_n​,C_k​)⋅P(x_2​∣x_3​,…,x_n​,C_k​)⋯P(x_n​∣C_k​)
$$
This becomes **intractable.**


## With Naive Assumption

Features become independent of class

So:
$$
P(x_i​∣x_j​,C_k​)=P(x_i​∣C_k​)
$$
this simpliflies to :

$$
P(X | C_k) = \prod_{1}^{n} P(x_i | C_k)
$$

### Example: Bernoulli Naive Bayes

Note : $ x \in \{0,1\}$ 
<br>

This is example is for **Bernoulli Naive Bayes** and not **Gaussian Naive Bayes** which would take continous data

```python
import numpy as np

# training data

X = np.array([
    [1,0],
    [1,1],
    [0,0],
    [0,1]
])

y = np.array([1,1,0,0]) #Labels

classes = np.unique(y) # 0 and 1
```

```python
# finding priors P(Classes)
c0 = 0 # class 0
c1 = 1 # class 1


p0 = np.mean(y == c0) #  = 2/4

p1 = np.mean( y == c1)
```

```python
# Seperate data for class
X0 =  X[y==c0]
X1 =  X[y==c1]
```

```python
# Likelihood
# we use laplace smoothing to avoid zero


p_x0 = (X0.sum(axis = 0)+ 1) / len((X0)+2) # P(X0 | C)
p_x1 = (X1.sum(axis = 0)+ 1) / len((X1)+2) # P(X1 | C)
```

Laplace smoothing:

$P(x_i|y)=\frac{count+1}{count(y)+V}$

<br>
V = number of possible values ; here it is 2

```python
def predict(x):
    prob0 = p0
    prob1 = p1

    for i in range (len(x)):
        if x[i] == 1:
            prob0 *= p_x0[i]
            prob1 *= p_x1[i]
        
        else:
            prob0 *= (1 - p_x0[i])
            prob1 *= (1 - p_x1[i])
    return 0 if prob0 > prob1 else 1
```

```python
predict([0,1])
```

```python
predict([1,1])
```

Note : $ x \in \{0,1\}$ , This is example is for **Bernoulli Naive Bayes** and not **Gaussian Naive Bayes** which would take continous data
